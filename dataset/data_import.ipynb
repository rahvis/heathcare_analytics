```python
import pandas as pd
```

```python
# Left lower extremity
GRF_F_V_PRO_left = pd.read_csv(r'GRF_F_V_PRO_left.csv')
#GRF_F_V_RAW_left = pd.read_csv(r'GRF_F_V_RAW_left.csv')

GRF_F_AP_PRO_left = pd.read_csv(r'GRF_F_AP_PRO_left.csv')
#GRF_F_AP_RAW_left = pd.read_csv(r'GRF_F_AP_RAW_left.csv')

GRF_F_ML_PRO_left = pd.read_csv(r'GRF_F_ML_PRO_left.csv')
#GRF_F_ML_RAW_left = pd.read_csv(r'GRF_F_ML_RAW_left.csv')

GRF_COP_AP_PRO_left = pd.read_csv(r'GRF_COP_AP_PRO_left.csv')
#GRF_COP_AP_RAW_left = pd.read_csv(r'GRF_COP_AP_RAW_left.csv')

GRF_COP_ML_PRO_left = pd.read_csv(r'GRF_COP_ML_PRO_left.csv')
#GRF_COP_ML_RAW_left = pd.read_csv(r'GRF_COP_ML_RAW_left.csv')

# Right lower extremity
GRF_F_V_PRO_right = pd.read_csv(r'GRF_F_V_PRO_right.csv')
#GRF_F_V_RAW_right = pd.read_csv(r'GRF_F_V_RAW_right.csv')

GRF_F_AP_PRO_right = pd.read_csv(r'GRF_F_AP_PRO_right.csv')
#GRF_F_AP_RAW_right = pd.read_csv(r'GRF_F_AP_RAW_right.csv')

GRF_F_ML_PRO_right = pd.read_csv(r'GRF_F_ML_PRO_right.csv')
#GRF_F_ML_RAW_right = pd.read_csv(r'GRF_F_ML_RAW_right.csv')

GRF_COP_AP_PRO_right = pd.read_csv(r'GRF_COP_AP_PRO_right.csv')
#GRF_COP_AP_RAW_right = pd.read_csv(r'GRF_COP_AP_RAW_right.csv')

GRF_COP_ML_PRO_right = pd.read_csv(r'GRF_COP_ML_PRO_right.csv')
#GRF_COP_ML_RAW_right = pd.read_csv(r'GRF_COP_ML_RAW_right.csv')

# Annotations and Metadata
GRF_Annotation = pd.read_csv(r'GRF_metadata.csv')
```

```python
GRF_F_V_PRO = pd.concat([GRF_F_V_PRO_left, GRF_F_V_PRO_right], axis = 0)
```

```python
GRF_F_AP_PRO = pd.concat([GRF_F_AP_PRO_left, GRF_F_AP_PRO_right], axis = 0)
```

```python
GRF_F_ML_PRO = pd.concat([GRF_F_ML_PRO_left, GRF_F_ML_PRO_right], axis = 0)
```

```python
GRF_COP_AP_PRO = pd.concat([GRF_COP_AP_PRO_left, GRF_COP_AP_PRO_right], axis = 0)
```

```python
GRF_COP_ML_PRO = pd.concat([GRF_COP_ML_PRO_left, GRF_COP_ML_PRO_right], axis = 0)
```

```python
GRF = pd.concat([GRF_F_V_PRO, GRF_F_AP_PRO, GRF_F_ML_PRO, GRF_COP_AP_PRO, GRF_COP_ML_PRO], axis = 0)
```

```python
GRF
```

```python
merged = pd.merge(GRF, GRF_Annotation, on="SESSION_ID")
```

```python
merged.to_csv("merged.csv", mode = 'w', index=False)
```
